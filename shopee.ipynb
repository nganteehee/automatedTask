```python
df.drop_duplicates(subset=['name','adress', 'image_url', 'menu' ], inplace=True)
df = df[~df.href.isna()]
```

```python
df.shape
```

```python
import re
```

```python
def check_formant(string):
    pattern = re.compile('.*:.*,')
    result = pattern.search(string)
    if result:
        return True
    else:
        return False
```

```python
df.shape
```

```python
df = df[df['menu'].apply(check_formant)]
```

```python
def safe_split(string):
    parts = string.split(':', 1) #max split = 1 to get exactly 2 parts
    if len(parts) ==2:
        return parts[0], parts[1]
    else:
        return None




def extract_name_price(string):
    ls = []
    # pattern = re.compile('VẤN ĐỀ')
    for i in string.strip('{}').split(","):
        # if pattern.search(i):
        #     continue 
        print(i)
        if safe_split(i):
            dish, pri = safe_split(i)
            print(dish, pri)
            ls.append((dish, pri))
            # item.append(dish)
            # price.append(pri)
        else:
            continue
    return ls
```

```python
df['menu_item'] = df['menu'].apply(extract_name_price)
df_exploded = df.explode('menu_item')
```

```python
df_exploded[['item', 'price']] = pd.DataFrame(df_exploded['menu_item'].tolist(), index = df_exploded.index)
```

```python
df_exploded['item']
```

```python
df_exploded['item'] = df_exploded['item'].str.replace("'", "")
df_exploded['price'] = df_exploded['price'].str.replace("'", "")
df_exploded['price'] = df_exploded['price'].str.replace("đ", "")
df_exploded['price'] = df_exploded['price'].astype(float)
```

```python
pattern = 'VẤN ĐỀ'
pattern_2 = ':'
df_exploded = df_exploded[~df_exploded['price'].str.contains(pattern, regex=True, na=False)]
df_exploded = df_exploded[~df_exploded['price'].str.contains(pattern_2, regex=True, na=False)]
```
